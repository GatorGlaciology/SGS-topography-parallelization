```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.gridspec as gridspec
```

```python
res = pd.read_csv('Test/res_200_2000.csv')
```

```python
res[res.execution == 'Sequential']['res']
```

```python
l = len(res)
res_s_t = res[res.execution == 'Sequential']['tot time']#[:10]
res_p_t = res[res.execution == 'Parallel']['tot time']#[:10]
num_pts = res[res.execution == 'Sequential']['num missing pts']#[:10]
```

```python
gs = gridspec.GridSpec(2, 2)
fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(5)
fig.suptitle('Horizontally stacked subplots')

ax0

ax1 = plt.subplot(gs[1, 0])
ax1.plot(num_pts, res_s_t, label ='Serial', marker='^', markerfacecolor='black', color = 'blue')
ax1.plot(num_pts, res_p_t, label ='Parallel', marker='^', markerfacecolor='black', color = 'orange')

ax1.set_xlabel("Number of Simulated Points")
ax1.set_ylabel("SGS Run Time (s)")
ax1.legend()
ax1.set_title('Run Time vs Number of Simulated Points')

ax2 = plt.subplot(gs[1, 1])
ax2.plot(num_pts, speedup, label ='Serial', marker='^', markerfacecolor='black', color = 'blue')

ax2.set_xlabel("Number of Simulated Points")
ax2.set_ylabel("Speed-up")
ax2.set_title('Speed-up vs Number of Simulated Points')
plt.show()
```
