```python
s = 0.7

fig, axx = plt.subplots(2, 3, figsize=(12,10))

im = axx[0,0].scatter(df_PIG_GT1[x], df_PIG_GT1[y], c=df_PIG_GT1[z], marker='.', s=s, cmap='gist_earth')
axx[0,0].set_title('PIG \'Ground Truth\' 1')

axx[0,0].set_xlabel('X (m)')
axx[0,0].set_ylabel('Y (m)')

im = axx[1,0].scatter(df_PIG_GT2[x], df_PIG_GT2[y], c=df_PIG_GT2[z], marker='.', s=s, cmap='gist_earth')
axx[1,0].set_title('PIG \'Ground Truth\' 1')

axx[1,0].set_xlabel('X (m)')
axx[1,0].set_ylabel('Y (m)')

axx[0,1].scatter(df_vert_path[x], df_vert_path[y], c=df_vert_path[z],marker='.',s=s, cmap='gist_earth')
axx[0,1].set_title('Vertical Path')
axx[0,1].set_xlabel('X (m)')
axx[0,1].yaxis.set_visible(False)

axx[0,2].scatter(df_horiz_path[x], df_horiz_path[y], c=df_horiz_path[z],marker='.',s=s, cmap='gist_earth')
axx[0,2].set_title('Horizontal Path')
axx[0,2].set_xlabel('X (m)')
axx[0,2].yaxis.set_visible(False)

axx[0,3].scatter(df_grid_path[x], df_grid_path[y], c=df_grid_path[z],marker='.',s=s, cmap='gist_earth')
axx[0,3].set_title('Grid Path')
axx[0,3].set_xlabel('X (m)')
axx[0,3].yaxis.set_visible(False)

axx[1,1].scatter(df_posdiag_path[x], df_posdiag_path[y], c=df_posdiag_path[z],marker='.',s=0.5, cmap='gist_earth')
axx[1,1].set_title('Positive Diagonal Path')
axx[1,1].set_xlabel('X (m)')
axx[1,1].yaxis.set_visible(False)

cbar = plt.colorbar(im, orientation="vertical", ax=axx[:,2], ticks=np.linspace(-470, 470, 11))
cbar.set_label('Bed (m)', rotation=270, labelpad=10)

plt.show()
fig.savefig('/Users/nathanschoedl/Desktop/Research_Figures/PIG_DIFF.png', bbox_inches='tight')
```
